# DATA ENCODING WITH PYTHON :- 

### 1step - LABEL ENCODER :- 

```python
phone_bb = pd.DataFrame([
        ["Phone 1", "Blue", 2018, 500],
        ["Phone 2", "Black", 2016, 450],
        ["Phone 3", "Blue", 2017, 600],
        ["Phone 4", "Grey", 2018, 800]])

phone_bb.columns = ["Phone", "Color", "Year", "Price"]
phone_bb
```

```python
from sklearn.preprocessing import LabelEncoder

le_pho = LabelEncoder()
le_col = LabelEncoder()

phone_bb["Phone_nc"] = le_pho.fit_transform(phone_bb["Phone"])
phone_bb["Color_nc"] = le_col.fit_transform(phone_bb["Color"])
```

```python
phone_bb
```

### 2step -  One-Hot Encoding :-  "convert categorical value using one hot enconding" 

```python
from sklearn.preprocessing import OneHotEncoder

l_pho = OneHotEncoder()
l_col = OneHotEncoder()

l_pho = l_pho.fit_transform(phone_bb["Phone_nc"].values.reshape(-1,1)).toarray()
l_col = l_col.fit_transform(phone_bb["Color_nc"].values.reshape(-1,1)).toarray()
```

```python
l_pho
```

### 3step - Adding the ohe'd feature to the original dataset :- "PHONE AND COLOR"

```python


phone_ohe = pd.DataFrame(l_pho,columns=["Phone_"+str(int(i)) for i in range(l_pho.shape[1])])
phone_bb = pd.concat([phone_bb,phone_ohe],axis=1)
```

```python
color_ohe = pd.DataFrame(l_col,columns=["Color_"+str(int(i)) for i in range(l_col.shape[1])])
phone_bb = pd.concat([phone_bb,color_ohe],axis=1)
```

```python
phone_bb.drop(["Phone","Color","Phone_nc","Color_nc"],axis=1,inplace=True)
```

```python
phone_bb
```

## Data Encoding Practice : - 

```python
dab = pd.DataFrame({"fruits" : ["Apple","Mango","banana","chiku"],"price" : [10,20,40,20]})
dab
```

```python
from sklearn.preprocessing import LabelEncoder

fruits_le = LabelEncoder()

dab["f"] = fruits_le.fit_transform(dab["fruits"]) 
```

```python
dab
```

```python
from sklearn.preprocessing import OneHotEncoder

fruits_ohe = OneHotEncoder()

fruits_ohe = fruits_ohe.fit_transform(dab["f"].values.reshape(-1,1)).toarray()
```

```python
fruits_ohe
```

```python
fruits_final = pd.DataFrame(fruits_ohe,columns=['fruits_'+str(int(i)) for i in range(fruits_ohe.shape[1])])
```

```python
dab = pd.concat([dab,fruits_final],axis=1)
```

```python
dab
```

```python
dab.drop(['fruits','f'],axis=1,inplace=True)
```

```python
dab
```

```python
phone_qq = pd.DataFrame([
        ["Phone 1", "Blue", 2018, 500],
        ["Phone 2", "Black", 2016, 450],
        ["Phone 3", "Blue", 2017, 600],
        ["Phone 4", "Grey", 2018, 800]])

phone_qq.columns = ["Phone", "Color", "Year", "Price"]
phone_qq
```
